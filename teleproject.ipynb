# Analyzing the Percentage column

```python
df.head(2)                                                                                #to see the top 2 record
```

```python

# Convert 'Churn Label' to boolean values (True for 'Yes', False for 'No')
df['Churn Label 1'] = df['Churn Label'].map({'Yes': True, 'No': False})

# Create the crosstab
k = pd.crosstab(df.City, df['Churn Label 1'], margins=True)

# Creating new column 'percentage_churn'
# 'True' represents the count of churned customers, so we use it to calculate percentage churn
k['percentage_churn'] = (k[True] / k['All']) * 100

print(k.head())
```

```python
k.info()                                                                      #to see the info
```

```python
k.sort_values(by='percentage_churn',ascending=False).head()                     #to see the descend value and top record of percentage column 
```

```python
k.sort_values(by='percentage_churn',ascending=False).tail()                  #to see the descend value and bottom record of percentage column 
```

```python
k.sort_values(by='percentage_churn',ascending=False).sample()           #to see the descend value and sample record of percentage column     
```
